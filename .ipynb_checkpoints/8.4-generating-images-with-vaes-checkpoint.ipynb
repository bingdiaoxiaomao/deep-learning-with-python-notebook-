### Variational autoencoders

```python
from keras import backend as K
K.clear_session()
```

```python
import keras
from keras import layers
from keras import backend as K
from keras.models import Model
import numpy as np
```

```python
img_shape = (28,28,1)
batch_size = 16
latent_dim = 2
input_img = layers.Input(shape=img_shape)
x = layers.Conv2D(filters=32,kernel_size=(3,3),padding="same",strides=(1,1),activation="relu")(input_img)

x = layers.Conv2D(filters=64,kernel_size=(3,3),padding="same",strides=(2,2),activation="relu")(x)

x = layers.Conv2D(filters=64,kernel_size=(3,3),padding="same",strides=(1,1),activation="relu")(x)

x = layers.Conv2D(filters=64,kernel_size=(3,3),padding="same",strides=(1,1),activation="relu")(x)

shape_before_flattening = K.int_shape(x)

x = layers.Flatten()(x)
x = layers.Dense(units=32)(x)
z_mean = layers.Dense(units=latent_dim)(x)
z_log_var = layers.Dense(units=latent_dim)(x)
```

```python
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0],latent_dim),mean=0.,stddev=1.)
    return z_mean + K.exp(z_log_var)*epsilon
z = layers.Lambda(sampling)([z_mean,z_log_var])
```

```python
decoder_input = layers.Input(shape=K.int_shape(z)[1:])

x = layers.Dense(units=32,activation="relu")(decoder_input)
x = layers.Dense(units=np.prod(shape_before_flattening[1:]),activation='relu')(x)
x = layers.Reshape(target_shape=shape_before_flattening[1:])(x)

x = layers.Conv2DTranspose(filters=64,kernel_size=(3,3),padding="same",strides=(1,1),activation="relu")(x)
x = layers.Conv2DTranspose(filters=64,kernel_size=(3,3),padding="same",strides=(1,1),activation="relu")(x)
x = layers.Conv2DTranspose(filters=64,kernel_size=(3,3),padding="same",strides=(2,2),activation="relu")(x)
x = layers.Conv2DTranspose(filters=32,kernel_size=(3,3),padding="same",strides=(1,1),activation="relu")(x)
x = layers.Conv2DTranspose(filters=1,kernel_size=(3,3),padding="same",strides=(1,1),activation="sigmoid")(x)

decoder = Model(inputs=decoder_input,outputs=x)
z_decoded = decoder(z)
```

```python
print(decoder.summary())
```

```python
class CustomVariationalLayer(keras.layers.Layer):
    
    def vae_loss(self,x,z_decoded):
        x = K.flatten(x)
        z_decoded = K.flatten(z_decoded)
        xent_loss = keras.metrics.binary_crossentropy(x,z_decoded)
        kl_loss = -5e-4*K.mean(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var),axis=-1)
        return K.mean(xent_loss + kl_loss)
    
    def call(self,inputs):
        x = inputs[0]
        z_decoded = inputs[1]
        loss = self.vae_loss(x,z_decoded)
        self.add_loss(loss,inputs=inputs)
        return x
    
y = CustomVariationalLayer()([input_img,z_decoded])
```

```python
from keras.datasets import mnist

vae = Model(input_img, y)
vae.compile(optimizer='rmsprop', loss=None)
vae.summary()

# Train the VAE on MNIST digits
(x_train, _), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255.
x_train = x_train.reshape(x_train.shape + (1,))
x_test = x_test.astype('float32') / 255.
x_test = x_test.reshape(x_test.shape + (1,))

vae.fit(x=x_train, y=None,
        shuffle=True,
        epochs=10,
        batch_size=batch_size,
        validation_data=(x_test, None))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import norm

# Display a 2D manifold of the digits
n = 15  # figure with 15x15 digits
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
# Linearly spaced coordinates on the unit square were transformed
# through the inverse CDF (ppf) of the Gaussian
# to produce values of the latent variables z,
# since the prior of the latent space is Gaussian
grid_x = norm.ppf(np.linspace(0.05, 0.95, n))
grid_y = norm.ppf(np.linspace(0.05, 0.95, n))

for i, yi in enumerate(grid_x):
    for j, xi in enumerate(grid_y):
        z_sample = np.array([[xi, yi]])
        z_sample = np.tile(z_sample, batch_size).reshape(batch_size, 2)
        x_decoded = decoder.predict(z_sample, batch_size=batch_size)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        figure[i * digit_size: (i + 1) * digit_size,
               j * digit_size: (j + 1) * digit_size] = digit

plt.figure(figsize=(10, 10))
plt.imshow(figure, cmap='Greys_r')
plt.show()
```
