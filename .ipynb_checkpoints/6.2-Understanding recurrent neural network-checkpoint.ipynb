```python
from keras.models import Sequential
from keras.layers import Embedding,SimpleRNN
```

默认设置情况下，仅仅返回RNN的最后一个输出

```python
model = Sequential()
model.add(Embedding(input_dim=10000,output_dim=32))
model.add(SimpleRNN(units=32))
model.summary()
```

通过设置return_sequences=True，使得中间层返回RNN的整个输出序列，最后一层仅返回最后序列的最后一个值

```python
model = Sequential()
model.add(Embedding(input_dim=10000,output_dim=32))
model.add(SimpleRNN(units=32,return_sequences=True))
model.summary()
```

```python
model = Sequential()
model.add(Embedding(input_dim=10000,output_dim=32))
model.add(SimpleRNN(units=32,return_sequences=True))
model.add(SimpleRNN(units=32,return_sequences=True))
model.add(SimpleRNN(units=32,return_sequences=True))
model.add(SimpleRNN(units=32))
model.summary()
```

```python
from keras.datasets import imdb
from keras.preprocessing import sequence
```

```python
max_features = 10000
maxlen = 500
batch_size = 32
```

```python
print('Loading data...')
(input_train,y_train),(input_test,y_test) = imdb.load_data(num_words=max_features)
print(len(input_train),'train sequences')
print(len(input_test),'test sequences')
```

```python
print('Pad sequences (samples x time)')
input_train = sequence.pad_sequences(sequences=input_train,maxlen=maxlen)
input_test = sequence.pad_sequences(sequences=input_test,maxlen=maxlen)
print('input_train shape {}'.format(input_train.shape))
print('input_test shape {}'.format(input_test.shape))
```

```python
from keras.layers import Dense
model = Sequential()
model.add(Embedding(input_dim=max_features,output_dim=32))
model.add(SimpleRNN(units=32))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
```

```python
history = model.fit(input_train,
                    y_train,
                    epochs=15,
                    batch_size=64,
                    validation_split=0.2)
```

```python
del model
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
epochs = history.epoch
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.plot(epochs,acc,'o-',color='r',label='Training')
plt.plot(epochs,val_acc,'-D',color='b',label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.title('Training and validation accuracy')

plt.subplot(122)
plt.plot(epochs,loss,'o-',color='r',label='Training')
plt.plot(epochs,val_loss,'-D',color='b',label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.title('Training and validation loss')
plt.show()
```

### A concrete LSTM example in keras

```python
from keras.layers import LSTM

model = Sequential()
model.add(Embedding(input_dim=max_features,output_dim=32))
model.add(LSTM(units=32))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
model.summary()
```

```python
history = model.fit(input_train,
                    y_train,
                    epochs=15,
                    batch_size=64,
                    validation_split=0.2)
```

```python
epochs = history.epoch
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.plot(epochs,acc,'o-',color='r',label='Training')
plt.plot(epochs,val_acc,'-D',color='b',label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.title('Training and validation accuracy')

plt.subplot(122)
plt.plot(epochs,loss,'o-',color='r',label='Training')
plt.plot(epochs,val_loss,'-D',color='b',label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.title('Training and validation loss')
plt.show()
```
